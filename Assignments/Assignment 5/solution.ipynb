## Part 1 – `myHashMap` using Separate Chaining

You were asked to complete these four methods in `myHashMap.java`:

1. `remove(K)`
2. `remove(K, V)`
3. `replace(K, V)`
4. `replace(K, V, V)`

The implementation uses **separate chaining**, so each bucket stores a linked list of nodes.
That means each method must:

- compute the bucket index
- traverse the linked list in that bucket
- update links carefully when removing or replacing


### Question 1 – `remove(K)`

**Goal:** Remove the entry with the given key and return its value.

**Idea:** Traverse the linked list in the correct bucket. If the key is found, unlink that node.

- If it is the first node in the bucket, update the bucket head.
- Otherwise, connect the previous node to the next node.
- Decrease `size`.
- Return the removed value.
- If not found, return `null`.


```java
public V remove(K key) {
    int index = getBucketIndex(key);
    HashNode<K, V> head = bucket.get(index);
    HashNode<K, V> prev = null;

    while (head != null) {
        if (Objects.equals(head.key, key)) {
            V removedValue = head.value;

            if (prev == null) {
                bucket.set(index, head.next);
            } else {
                prev.next = head.next;
            }

            size--;
            return removedValue;
        }

        prev = head;
        head = head.next;
    }

    return null;
}
```

### Question 2 – `remove(K, V)`

**Goal:** Remove the entry only if both the key and value match.

**Idea:** This is similar to `remove(K)`, but now both conditions must be true.

- If the key exists but the value does not match, do not remove it.
- Return `true` only when a matching pair is actually removed.
- Otherwise return `false`.


```java
public boolean remove(K key, V val) {
    int index = getBucketIndex(key);
    HashNode<K, V> head = bucket.get(index);
    HashNode<K, V> prev = null;

    while (head != null) {
        if (Objects.equals(head.key, key) && Objects.equals(head.value, val)) {
            if (prev == null) {
                bucket.set(index, head.next);
            } else {
                prev.next = head.next;
            }

            size--;
            return true;
        }

        prev = head;
        head = head.next;
    }

    return false;
}
```

### Question 3 – `replace(K, V)`

**Goal:** Replace the value for a key only if that key already exists.

**Idea:** Traverse the correct bucket and look for the key.

- If found, store the old value.
- Update the node’s value.
- Return the old value.
- If the key is not found, return `null`.

This matches the behavior of Java’s `HashMap.replace(key, value)`.


```java
public V replace(K key, V val) {
    int index = getBucketIndex(key);
    HashNode<K, V> head = bucket.get(index);

    while (head != null) {
        if (Objects.equals(head.key, key)) {
            V oldValue = head.value;
            head.value = val;
            return oldValue;
        }
        head = head.next;
    }

    return null;
}
```

### Question 4 – `replace(K, V, V)`

**Goal:** Replace the value only if the key exists **and** its current value equals `oldVal`.

**Idea:**

- Search for the key in the correct bucket.
- Check whether the current value equals `oldVal`.
- If yes, replace it with `newVal` and return `true`.
- Otherwise return `false`.

This version can also reuse the previous `replace(K, V)` method, avoiding duplication and giving a consistent of behavior.


```java
public boolean replace(K key, V oldVal, V newVal) {
    int index = getBucketIndex(key);
    HashNode<K, V> head = bucket.get(index);

    while (head != null) {
        if (Objects.equals(head.key, key) && Objects.equals(head.value, oldVal)) {
            replace(key, newVal);
            return true;
        }
        head = head.next;
    }

    return false;
}
```

### Question 5 – `getAverage(HashMap<Integer, Integer> map, int[] array)`

**Goal:** Compute the average of the values whose keys appear in both the array and the map.

**Idea:**

- Loop through the array.
- For each number, check whether it is a key in the map.
- If yes, add the corresponding value to the sum and increase the count.
- Return `sum / count`.

If no common keys exist, `sum / count` becomes `0.0 / 0`, which gives `NaN`, and that is exactly what the assignment wants.


```java
public double getAverage(HashMap<Integer, Integer> map, int[] array) {
    double sum = 0.0;
    int count = 0;
    int num;

    for (int i = 0; i < array.length; i++)  { 
        num = array[i];
        if (map.containsKey(num)) {
            sum += map.get(num);
            count++;
        }
    }

    return sum / count;
}
```

### Question 6 – `odd(HashMap<Integer, String> map)`

**Goal:** Return an `ArrayList` containing the values associated with odd keys.

**Idea:**

- Iterate through the keys of the map using `keySet()`.
- Check whether the key is odd using `key % 2 != 0`.
- If yes, add the corresponding value to the result list.


```java
public ArrayList<String> odd(HashMap<Integer, String> map) {
    ArrayList<String> result = new ArrayList<>();

    for (Integer key : map.keySet()) {
        if (key % 2 != 0) {
            result.add(map.get(key));
        }
    }

    return result;
}
```

### Question 7 – `twoSums(int[] numbers, int k)`

**Goal:** Count how many pairs of numbers have difference `k`, in **O(n)** time.

**Idea:**

- First store all numbers in a `HashSet`.
- Then loop through the array again.
- For each number `x`, check whether `x + k` exists.
- If yes, that gives one valid pair.

This is efficient because `HashSet.contains(...)` is average-case constant time.
So the total time is O(n).


```java
public int twoSums(int[] numbers, int k) {
    HashSet<Integer> seen = new HashSet<>();
    int count = 0;

    for (int num : numbers) {
        seen.add(num);
    }

    for (int num : numbers) {
        if (seen.contains(num + k)) {
            count++;
        }
    }

    return count;
}
```
