# Designing a Hybrid Hash–Tree Index in Java

## Overview

In this assignment, you will design and implement a hybrid data structure in Java that combines **hashing** and **tree-based organization** in order to manage collisions in a more structured way.

Hash tables typically provide very efficient average-case performance, often described as **O(1)** for insertion, search, and deletion under appropriate assumptions. However, when collisions accumulate in the same bucket, performance may deteriorate depending on how those collisions are handled.

The purpose of this project is to explore a design that combines two important ideas from this course:

- **Hashing**, for fast average-case access
- **Binary Search Trees (BSTs)**, for structured organization within buckets

Instead of resolving collisions with a linked list, each bucket in your hash table will store its entries using a **Binary Search Tree**. In addition, your structure must support **rehashing** when the load factor exceeds a chosen threshold.

This project is intended to move beyond basic implementation. You are expected not only to build a correct solution, but also to reason carefully about its behavior, limitations, and performance.

## Problem Description

Design and implement a class named `HybridHashMap` that supports the following public operations:

- `public void put(int key, int value)`  
  Inserts a new key-value pair into the structure, or updates the value if the key already exists.

- `public int get(int key)`  
  Returns the value associated with the key if the key is present. If the key is not present, return `-1`.

- `public void remove(int key)`  
  Removes the key and its associated value from the structure if the key exists.

Your implementation must use:

- a **hash table** as the top-level structure
- a **Binary Search Tree (BST)** inside each bucket to store colliding entries

## Design Requirements

### 1. Hash Table Structure

- Use an array of buckets.
- Each key must be mapped to a bucket using a hash function.
- The number of buckets must be tracked as part of the state of the data structure.

---

### 2. Collision Handling with BSTs

- Each bucket must store its elements using a **Binary Search Tree**.
- You must implement the BST logic yourself.
- You may assume that keys are comparable.

---

### 3. Required BST Operations

Your BST implementation must support the operations needed to make the hash map function correctly, including:

- insertion
- search
- deletion

---

### 4. Rehashing (Mandatory)

Your structure must dynamically resize when it becomes too full.

You must choose a load factor threshold (for example, `0.75`). When the load factor exceeds that threshold:

- create a new table with larger capacity, typically double the current number of buckets
- recompute bucket positions using the new table size
- reinsert all existing elements into the new table

Your implementation must include a clearly defined rehashing process.

---

### 5. Storage Constraints

- Do **not** use Java's built-in `HashMap`, `TreeMap`, `HashSet`, `TreeSet`, or similar collection classes as the underlying storage mechanism.
- Do **not** use built-in map or set classes to replace the required logic.
- The goal of this project is to implement the relevant data structure logic yourself.

Your design should aim for the following:

You are **not** required to implement a balanced BST. However, part of the project is to analyze the consequences of that decision.
