## Question 4 — Heapify and Priority Queue Sorting (30 points)

In this question, you will work with a priority queue implemented using a **min-heap**.

The goal is to understand how **heapify** builds a heap efficiently from an existing sequence of values,
and how a priority queue can then be used to sort a sequence.


## How Heapify Connects to What We Saw in Class

In class, we described heap construction as a **bottom-up process**: we start with small heaps and combine them into larger heaps.

In the array implementation, we use the same idea in a practical way.

A leaf is already a valid heap by itself.
So when a binary heap is stored in an array of length `n`:

- all indices greater than `(n / 2) - 1` are leaves
- all indices from `0` up to `(n / 2) - 1` are non-leaves

This means that all leaves already satisfy the heap property.
So to build the heap, we only need to process the non-leaf nodes.

We begin at the **last non-leaf node**, whose index is:

```java
(n / 2) - 1
```

Then we move backward toward index `0`.

At each index:

- the left and right subtrees have already been processed
- so they are already valid heaps
- the only possible issue is that the current node may not be in the correct position relative to its children

So for each non-leaf node, we perform a **bubble-down** operation if needed.

This is the array-based way to implement the same bottom-up heap construction idea seen in class.


### Why this works

At any point in this process, when considering a node, its children can be treated as roots of valid heaps. The only possible violation is at the node itself, which can be corrected by moving it downward until the heap property is restored.

### Why this is efficient

Because each element is moved only as far as necessary, and most elements are near the bottom of the tree, this bottom-up construction runs in **O(n)** time. This is more efficient than inserting elements one by one into an empty heap, which would take **O(n log n)** time.


## How We Index a Heap in an Array

If a heap is stored in an array:

- left child of index `i` is at `2*i + 1`
- right child of index `i` is at `2*i + 2`
- parent of index `i` is at `(i - 1) / 2`

Because of this indexing scheme, the leaves are all grouped at the end of the array.
That is why `(n / 2) - 1` is the last non-leaf node.


## Example Idea 

Suppose the array is:

```java
[7, 3, 9, 1, 6, 8, 2]
```

Tree representation:

```text
         7
       /   \
      3     9
     / \   / \
    1   6 8   2
```

The array length is `7`, so the last non-leaf node is:

```java
(7 / 2) - 1 = 2
```

So indices `3, 4, 5, 6` are leaves.
That means those nodes are already valid heaps.

We now process indices `2`, then `1`, then `0`.

### Step 1: process index 2

```text
         7
       /    \
      3       9(r)
     /  \    /    \
  1(b) 6(b)  8(b)  2(b)
```

The children of `9` are already roots of valid heaps.
But `9` is larger than `2`, so the heap property is violated.
We bubble `9` down.

After correction:

```java
[7, 3, 2, 1, 6, 8, 9]
```

Now the whole subtree rooted at index `2` is also a valid heap.

### Step 2: process index 1

```text
         7
       /   \
   3(r)         2(b)
    /  \     /      \
1(b)   6(b)  8(b)    9(b)
```

Again, the children are already roots of valid heaps.
But `3` is larger than `1`, so we bubble `3` down.

After correction:

```java
[7, 1, 2, 3, 6, 8, 9]
```

Now the subtree rooted at index `1` is a valid heap.

### Step 3: process index 0

```text
         7(r)
       /       \
   1(b)          2(b)
    /   \       /   \
3(b)    6(b)    8(b)   9(b)
```

Both subtrees are already heaps.
The only possible problem is the root.
We bubble `7` down until the heap property is restored.

After correction:

```java
[1, 3, 2, 7, 6, 8, 9]
```

Now the whole tree is a valid min-heap.
