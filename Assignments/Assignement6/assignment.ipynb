## Assignment Overview

This assignment contains three short problems for which you will develop solutions. When solving these problems, utilize the **HashMap** and **PriorityQueue** classes from the Java Collections Framework as appropriate.

Two additional object classes are provided: **BloomFilter** and **CuckooHash**. You will complete the templated methods for these classes.

## Problem 1 — Hash Table (21 points)

Given two arrays **A** and **B**, return whether array **B** is a subset of array **A**.

**Example 1:**  
Input: `[1, 50, 55, 80, 90]`, `[55, 90]`  
Output: `true`

**Example 2:**  
Input: `[1, 50, 55, 80, 90]`, `[55, 90, 99]`  
Output: `false`


## Problem 2 — Priority Queue 1 (21 points)

Given an array **A** and an integer **k**, return the **kth maximum element** in the array.

**Example:**  
Input: `[1, 7, 3, 10, 34, 5, 8]`, `k = 4`  
Output: `7`


## Problem 3 — Priority Queue 2 (21 points)

Given two arrays **A** and **B** with `n` and `m` integers respectively, return all elements in **sorted order**.

**Example:**  
Input: `[4, 1, 5]`, `[3, 2]`  
Output: `1 2 3 4 5`


## Bloom Filter Hash Map (21 points)

Implement the following method:

- `contains`

Complete the `contains()` method. This method checks the bits in the bloom filter map for each of the `k` hash codes based on the given parameter.

- It returns **false** if the element is definitely not in the set.
- It returns **true** if the element is **most probably** in the set.


## Cuckoo Hash Map (16 points)
